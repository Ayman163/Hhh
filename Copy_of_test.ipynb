```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)
print("Dataset:")
print(df.head())
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
print("\nDataset after setting Date index:")
print(df.head())
plt.figure(figsize=(12,5))
plt.plot(df.index, df["Temperature (°C)"], marker="o", markersize=2)
plt.title("Daily Temperature Over One Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
average_temp = df["Temperature (°C)"].mean()
print("\nAverage Temperature:", round(average_temp, 2))
max_temp = df["Temperature (°C)"].max()
print("Highest Temperature:", max_temp)
min_temp = df["Temperature (°C)"].min()
print("Lowest Temperature:", min_temp)
max_date = df["Temperature (°C)"].idxmax()
print("Date of Highest Temperature:", max_date.date())
min_date = df["Temperature (°C)"].idxmin()
print("Date of Lowest Temperature:", min_date.date())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_excel(file_name)

print("Dataset:")
print(df.head())
X = df[["Hours"]]
y = df["Score"]
model = LinearRegression()
model.fit(X, y)
prediction = model.predict([[9]])

final_prediction = min(prediction[0], 100)

print("Predicted score:", round(final_prediction, 2))

print("Predicted score for 9 study hours:", round(prediction[0], 2))
plt.scatter(df["Hours"], df["Score"])
plt.plot(df["Hours"], model.predict(X))

plt.title("Study Hours vs Score")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name)
data.head()
print("Shape:", data.shape)
print("\nInfo:")
print(data.info())
print("\nStatistics:")
print(data.describe())
data = data.dropna()
avg_salary = data.groupby("City")["Salary"].mean()
print(avg_salary)
city_counts = data["City"].value_counts()
print(city_counts)
print("Max Salary:", data["Salary"].max())
avg_salary.plot(kind="bar")
plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Salary")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
data = pd.read_excel(file_name)
data.head()
data["Category"] = data["Salary"].apply(
lambda x: "High" if x > 1800 else "Low" )
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X = data[["Age"]]
y = data["Category"]
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.2
)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(predictions)
from sklearn.cluster import KMeans
X = data[["Age", "Salary"]]
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
data["Cluster"] = kmeans.labels_
print(data.head().to_string())
import matplotlib.pyplot as plt
cluster_counts = data["Cluster"].value_counts()
plt.pie(
cluster_counts,
labels=cluster_counts.index,
autopct='%1.1f%%')
plt.title("Employees Distribution by Cluster")
plt.show()
```
